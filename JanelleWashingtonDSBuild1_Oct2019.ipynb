```
# Make dataframe

df = pd.read_csv ('Lead_Testing_in_School_Drinking_Water_Sampling_and_Results__Most_Recently_Reported_Beginning_2016.csv')
```

```
# Look at dataframe

df.head()
```

```
# Look at types of data for each column in dataset

df.dtypes
```

```
# Statistics for numerical data in dataset

df.describe()
```

```
# Looking only at schools that did not have any buildings with lead-free plumbing

no_lead_free_plumming = df[df['Any Buildings with Lead-Free Plumbing?']=='No']
```

```
no_lead_free_plumming.head()
```

```
no_lead_free_plumming.count()
```

```
# Looking only at schools that had any buildings with lead-free plumbing

yes_lead_free_plumming = df[df['Any Buildings with Lead-Free Plumbing?']!='No']
```

```
yes_lead_free_plumming
```

```
yes_lead_free_plumming.count()
```

```
# Make pie chart to depict results

dfpie = df.groupby('Any Buildings with Lead-Free Plumbing?').count()
```

```
dfpie
```

```
dfpie.School.plot(kind='pie')
```

```
# Make bar chart to depict results

dfpie.School.plot(kind='bar')
```

```
# Make heat map to show schools that had highest concentration of lead in pipes by county

data = df.pivot_table(index='County', values='Number of Outlets, Result > 15 ppb')
sns.heatmap(data)
```

```
# Looking at schools that had any buildings with lead-free pipes by county

lead_free_plumming_county = yes_lead_free_plumming.groupby('County').count()
```

```
lead_free_plumming_county
```

```
# Looking at schools that did not have any buildings with lead-free pipes by county

no_lead_free_plumming_county = no_lead_free_plumming.groupby('County').count()
```
